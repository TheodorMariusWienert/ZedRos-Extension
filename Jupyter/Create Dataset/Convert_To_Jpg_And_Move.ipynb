```python
import cv2
import os 
```

```python
path_to_images="/home/theubuntu/Robot_Data_Hichem/dataset/data_png/"
destination_path_iamges="/home/theubuntu/yolov5/datasets/Dataset_Robots_Humans/images/all"
```

```python
src_images = os.listdir(path_to_images) #Moves images
for image_name in src_images:
    full_path_file= os.path.join(path_to_images, image_name)
    if os.path.isfile(full_path_file):
        #print(full_path_file)
        im = cv2.imread(full_path_file)
        full_des_path_file=os.path.join(destination_path_iamges, image_name)
        full_des_path_file=os.path.splitext(full_des_path_file)[0]+".jpg"
        #print(full_des_path_file)
        cv2.imwrite(full_des_path_file,im, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```
